```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
loan_data = pd.read_csv('train.csv')

# Display the first few rows of the dataset to understand its structure
loan_data.head()
```

#Bar Plot for Loan Status

Bar plots are used to display and compare the frequency, count, or other measures (such as mean, sum, etc.) of different categories. They are helpful for visualizing categorical data and for comparing values across categories.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Loan_Status', data=loan_data, palette='Set2')
plt.title('Loan Status Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.show()
```

The palette parameter in Seaborn functions is used to define the color scheme of the plot.

#Histogram for Loan Amount


Histograms are used to display the distribution of a numerical variable. They show how data is distributed across different ranges or bins. Histograms are useful for understanding the underlying distribution of the data, detecting patterns, and identifying potential outliers.

```python
plt.figure(figsize=(10, 5))
sns.histplot(loan_data['LoanAmount'].dropna(), kde=True, color='skyblue')
plt.title('Distribution of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

kde=True: This adds a kernel density estimate line to the histogram, providing a smoothed estimate of the data distribution.

#Box Plot for Loan Amount by Education

A box plot, also known as a box-and-whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. Box plots are particularly useful for identifying outliers, understanding the spread and skewness of the data, and comparing distributions between different groups.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Education', y='LoanAmount', data=loan_data, palette='Set3')
plt.title('Loan Amount by Education')
plt.xlabel('Education')
plt.ylabel('Loan Amount')
plt.show()
```

#Scatter Plot for Applicant Income vs Loan Amount

A scatter plot is a type of data visualization that is used to display the relationship between two numerical variables. Each point on the scatter plot represents an observation from the dataset, with the position determined by the values of the two variables.

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=loan_data, hue='Loan_Status', palette='Set1')
plt.title('Applicant Income vs Loan Amount')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

hue is an argument used to add a third dimension to the scatter plot by applying color coding based on a categorical variable.

#Heatmap of Correlation Matrix

A heatmap is a data visualization technique that displays the magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving a quick visual summary of information.

```python
# Select only numeric columns for correlation matrix
numeric_loan_data = loan_data.select_dtypes(include=['number'])

# Fill NaN values with the mean of each column
numeric_loan_data = numeric_loan_data.fillna(numeric_loan_data.mean())

# Calculate the correlation matrix
correlation_matrix = numeric_loan_data.corr()

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

#Count Plot for Gender

A count plot is used to visualize the count of observations in each categorical bin using bars. It is particularly useful when you want to display the frequency distribution of a categorical variable.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Gender', data=loan_data, palette='Set1')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#Pair Plot for Selected Features

A pair plot, also known as a scatterplot matrix, is a powerful tool for visualizing the relationships between multiple variables in a dataset.

```python
plt.figure(figsize=(10, 5))
sns.pairplot(loan_data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Status']], hue='Loan_Status', palette='Set1')
plt.show()
```

#Violin Plot for Loan Amount by Gender

A violin plot is a powerful tool for visualizing the distribution of data across different categories.

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x='Gender', y='LoanAmount', data=loan_data, palette='Set3')
plt.title('Loan Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Loan Amount')
plt.show()
```

#Box Plot for Applicant Income by Loan Status

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=loan_data, palette='Set2')
plt.title('Applicant Income by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Applicant Income')
plt.show()
```

#Dist Plot for Coapplicant Income

A distribution plot (dist plot) is used to visualize the distribution of a dataset, showing the frequency or density of data points across the range of values.

```python
plt.figure(figsize=(10, 5))
sns.histplot(loan_data['CoapplicantIncome'].dropna(), kde=True, color='purple')
plt.title('Distribution of Coapplicant Income')
plt.xlabel('Coapplicant Income')
plt.ylabel('Frequency')
plt.show()
```

#Strip Plot

A strip plot is a type of plot used to represent and compare the distribution of data points for different categories. It is particularly useful when you want to visualize the relationship between a categorical variable and a continuous variable.

```python
# Create a strip plot
plt.figure(figsize=(12, 8))
sns.stripplot(x='Education', y='LoanAmount', data=loan_data, jitter=True)
plt.title('Loan Amount Distribution by Education')
plt.xlabel('Education')
plt.ylabel('Loan Amount')
plt.show()
```

The jitter=True parameter is used to spread out the points to avoid overlap.

#Factor Plot

A factor plot (also known as a categorical plot in Seaborn) is a powerful tool for visualizing the relationship between a categorical variable and one or more numerical variables

```python
# Create a factor plot
plt.figure(figsize=(12, 8))
sns.catplot(x='Gender', y='LoanAmount', hue='Education', kind='bar', data=loan_data)
plt.title('Loan Amount Distribution by Gender and Education')
plt.xlabel('Gender')
plt.ylabel('Loan Amount')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.catplot(x='Married', y='LoanAmount', hue='Education', kind='bar', data=loan_data)
plt.title('Loan Amount Distribution by Marital Status and Education')
plt.xlabel('Marital Status')
plt.ylabel('Loan Amount')
plt.show()
```

Change kind='bar' to kind='point' for a point plot, kind='box' for a box plot, or kind='violin' for a violin plot.

#Cat Plot

The catplot can be used to visualize the relationship between categorical variables and numerical variables.

```python
# Create a catplot
plt.figure(figsize=(12, 8))
sns.catplot(x='Gender', y='LoanAmount', hue='Education', kind='bar', data=loan_data)
plt.title('Loan Amount Distribution by Gender and Education')
plt.xlabel('Gender')
plt.ylabel('Loan Amount')
plt.show()
```

#Styled Plot

```python
# Set the style and color palette
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")

# Create a styled catplot
plt.figure(figsize=(12, 8))
catplot = sns.catplot(x='Gender', y='LoanAmount', hue='Education', kind='bar', data=loan_data, palette=palette, height=6)
catplot.set_axis_labels("Gender", "Loan Amount")
catplot.set_titles("Loan Amount Distribution by Gender and Education")
catplot.despine(left=True)

# Customize the plot's title
plt.title('Styled Loan Amount Distribution by Gender and Education', fontsize=16, weight='bold')
plt.show()
```

#FacetGrid

A FacetGrid is used in data visualization to create multiple plots based on subsets of a dataset. It helps in understanding the relationship between variables across different subsets by creating a grid of plots, each corresponding to a different subset of the data.

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Create a FacetGrid for LoanAmount by Education and Gender
g = sns.FacetGrid(loan_data, col="Education", row="Gender", margin_titles=True)
g.map(plt.hist, "LoanAmount", bins=20, color="steelblue")

# Add titles and labels
g.fig.suptitle('Distribution of Loan Amount by Education and Gender', fontsize=16, weight='bold')
g.set_axis_labels("Loan Amount", "Frequency")
g.set_titles(col_template="{col_name} Education", row_template="{row_name} Gender")

# Adjust the layout
g.fig.tight_layout(rect=[0, 0, 1, 0.95])
```
